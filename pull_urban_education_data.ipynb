```python
# Calculate the admission rate
df_adm['admit_rate'] = df_adm['number_admitted'] / df_adm['number_applied']

# Calculate the admission rate
df_adm['yield_rate'] = df_adm['number_enrolled_total'] / df_adm['number_admitted']

# Create a boolean mask for rows where sex is 99
mask = df_adm['sex'] == 99

# Filter the DataFrame using the mask
df_adm = df_adm[mask]

df_adm
```

```python
url = "https://educationdata.urban.org/api/v1/college-university/ipeds/sfa-all-undergraduates/2021/"

response = urlopen(url)
data_sfa = loads(response.read())
df_sfa = pd.DataFrame(data_sfa['results'])
df_sfa
```

```python
# prompt: make a new df called df_sfa_short that has 1 row for each unique unitid in df_sfa, and contains:  the sum of average_amount, total_amount across all type_of_aid, and the percent_of_students where type_of_aid == 1

# Group by unitid and calculate the required aggregations
df_sfa_short = df_sfa.groupby('unitid').agg(
    sum_average_amount=('average_amount', 'sum'),
    sum_total_amount=('total_amount', 'sum'),
    percent_of_students=('percent_of_students', 'mean')
).reset_index()

df_sfa_short
```

# aid applicants

```python
from urllib.request import urlopen
from json import loads
import pandas as pd

# Define the base URL
base_url = "https://educationdata.urban.org/api/v1/college-university/scorecard/student-characteristics/"

# Define the columns you want to keep
cols_shnc = [
    'unitid',
    'lowincome_pct',
    'midincome_pct',
]

# Create an empty list to store the dataframes
dfs = []

# Loop through the years from 2011 to 2021
for year in range(2015,2016):
    # Construct the URL for the current year, including fields in query
    fields = ",".join(cols_shnc)
    url = f"{base_url}{year}/aid-applicants/?fields={fields}"

    # Open the URL and load the JSON data
    response = urlopen(url)
    data = loads(response.read())

    # Convert the JSON data to a Pandas DataFrame
    df = pd.DataFrame(data['results'])

    # Append the subsetted dataframe to the list
    dfs.append(df)

# Concatenate all the dataframes in the list
df_all_years = pd.concat(dfs, ignore_index=True)

# Display the combined dataframe
df_all_years
```

# Schools Poverty MEPS

```python
url = "https://educationdata.urban.org/api/v1/schools/meps/2020/"
```

```python
import pandas as pd
from urllib.request import urlopen
from json import loads

# Load data from the first URL
url_ccd = "https://educationdata.urban.org/api/v1/schools/ccd/directory/2020/"
response_ccd = urlopen(url_ccd)
data_ccd = loads(response_ccd.read())
df_ccd = pd.DataFrame(data_ccd['results'])

# Select desired columns from the first dataframe
cols_ccd = [
    "ncessch",
    "zip_location",
    "highest_grade_offered",
    "school_level",
]
df_ccd_subset = df_ccd[cols_ccd]

# Load data from the second URL
url_meps = "https://educationdata.urban.org/api/v1/schools/meps/2020/"
response_meps = urlopen(url_meps)
data_meps = loads(response_meps.read())
df_meps = pd.DataFrame(data_meps['results'])

# Select desired columns from the second dataframe
cols_meps = [
    "ncessch",
    "meps_poverty_pct",
]
df_meps_subset = df_meps[cols_meps]

# Merge the two dataframes on 'ncessch'
df_schools = pd.merge(df_ccd_subset, df_meps_subset, on='ncessch', how='left')

# Display the merged dataframe
df_schools.head()
```

```python
# prompt: subset to school level 3 and give me the count

# Assuming df_schools is the DataFrame created in the previous code.
# Filter for school level 3
df_schools = df_schools[df_schools['school_level'] == 3]

# Get the count of schools with level 3
count_level_3 = len(df_schools)

# Print the count
print(f"The number of schools with level 3 is: {count_level_3}")
```

# Combine

```python
# Merge the dataframes based on 'unitid'
df_merged = df_shnc_sub.merge(df_dir_sub, on='unitid', how='left')
df_merged = df_merged.merge(df_sfa_short_sub, on='unitid', how='left')
df_merged = df_merged.merge(df_adm_sub, on='unitid', how='left')

# Display the merged dataframe
df_merged
```

# Clean

```python
# prompt: remove rows where zip is missing

# Remove rows where 'zip' is missing
df_merged = df_merged.dropna(subset=['zip'])

# Remove rows where 'admit_rate' is 0 or missing
df_merged = df_merged[df_merged['admit_rate'].notna()]
df_merged = df_merged[df_merged['admit_rate'] != 0]

df_merged = df_merged.replace({np.nan: None})

df_merged
```
